```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt 
```

```python
Rent = 500.
Elec = 30.
Publicit = 100.
Kfix = Rent + Elec + Publicit

Ingredients = 50.
Misc = 15.
Kvar = Ingredients + Misc

Prices = np.arange(150, 400, 50) #per batch 

# model of work hours as a function of the number of batches
for i in range(1,9):
    if i == 1:
        hours = [10] * 4
    else:
        for j in range(0,4):
            hours.append(10*i)
```

```python
batches = np.arange(0,32)
Costs = Kfix + Kvar*batches
```

```python
fig = plt.figure()

ax1 = plt.gca()

ax1.plot(batches, Costs, label='Costs')

for price in Prices:
    ax1.plot(batches, price*batches, label='Income w Price: {}'.format(price))

ax1.legend(loc=2)
ax1.set_xlabel('Number of batches (per month)')
ax1.set_ylabel('Euros')

ax2 = ax1.twinx()    

ax2.plot(batches, hours, 'k--', label='Work hours')

ax2.legend(loc=1)
ax2.set_ylabel('Work Hours');
```
